# Lowercasing

```python
df['Resume']=df['Resume'].apply(lambda x:x.lower())
```

```python
df
```

# Removing Stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
sw_list=stopwords.words('english')
```

```python
df['Resume']=df['Resume'].apply(lambda x:[item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
df
```

# Removing Numbers

```python
df['Resume']=df['Resume'].apply(lambda x:' '.join([i for i in x.split() if not i.isdigit()]))
```

```python
df
```

# Removing Punctuation

```python
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))
```

```python
df['Resume']=df['Resume'].apply(remove_punctuation)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3,stratify=y)
```

```python
print(x_train.shape,x_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
x_train=tfidf.fit_transform(x_train['Resume']).toarray()
x_test=tfidf.transform(x_test['Resume'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
y_pred
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import pickle
model_pkl_file = "resumeclassifier.pkl"  

with open(model_pkl_file, 'wb') as file:  
    pickle.dump(rf, file)
```

```python
with open(model_pkl_file, 'rb') as file:  
    rf = pickle.load(file)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
def predict_role(text):
    # (Optional) If you have a clean_text() function, apply it
    # text = clean_text(text)

    # Transform input with TF-IDF
    vectorized_text = tfidf.transform([text])

    # Predict encoded label with RandomForest
    predicted_category = rf.predict(vectorized_text)

    # Decode to original role name
    predicted_category_name = le.inverse_transform(predicted_category)

    return predicted_category_name[0]
```
